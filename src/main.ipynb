**Import library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load dataset**

```python
df = pd.read_csv("../data/Military_Operations_Strategic new.csv")
print(df.head())
```

**Statistik Deskriptif**

```python
#hitung statistik
mean_val = df["Year"].mean()
median_val = df["Year"].median()
mode_val = df["Year"].mode()[0]

# variabilitas
range_val = df['Year'].max() - df['Year'].min()
variance_val = df['Year'].var()
std_val = df['Year'].std()

# table hasil statistik
table_hasil_statistik = pd.DataFrame({
  "Statistik": ["Mean", "Median", "Mode", "Range", "Variance", "Std Dev"],
  "Nilai": [mean_val, median_val, mode_val, range_val, variance_val, std_val]
})

print(table_hasil_statistik)
```

**Visualisasi Distribusi**

```python
# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df['Year'], kde=True)
plt.title("Histogram + Density")
plt.show()

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Year'])
plt.title("Boxplot")
plt.show()
```
